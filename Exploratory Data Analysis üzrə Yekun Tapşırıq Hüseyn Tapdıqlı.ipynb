```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sales_data.csv", parse_dates=['Date'])
```

### `Customers_Age` üzrə datanın ədədi ortası nədir?

```python
# your code goes here
df['Customer_Age'].mean()
```

Bu datanı (`Customer_Age`) <b>density (KDE)</b> və <b>box plot</b> olaraq ifadə edin:

```python
# your code goes here
fig, [ax1, ax2] = plt.subplots(2, 1, figsize=(12,12))
df['Customer_Age'].plot(kind='kde', figsize=(10,10), fontsize=10, ax=ax1 )
df.boxplot(column =['Customer_Age'], color='blue',fontsize=10, ax=ax2)
plt.show()
```

### `Order_Quantity` üzrə datanın ədədi ortası nədir??

```python
# your code goes here
df['Order_Quantity'].sum() / df['Order_Quantity'].count()
```

`Order_Quantity` üzrə datanın <b>histogram</b> və <b>box plot</b>-unu göstərin:

```python
# your code goes here
fig, [ax1, ax2] = plt.subplots(2, 1, figsize=(12,12))
df.hist('Order_Quantity', ax=ax1 )
df.boxplot(column =['Order_Quantity'], color='blue',fontsize=10, ax=ax2)
plt.show()
```

### Illər üzrə satış miqdarı nə qədərdir?

```python
# your code goes here
df.groupby('Year')['Order_Quantity'].sum()
```

Bunu <b>pie plot</b> şəklində göstərin (illər üzrə faizlərə bölməklə):

```python
# your code goes here
df.groupby('Year')['Order_Quantity'].sum().plot(kind='pie', figsize=(10,10), autopct='%0.2f%%', fontsize=10)
```

### Aylar üzrə satış miqdarı nə qədərdir?

```python
# your code goes here
df.groupby('Month')['Order_Quantity'].sum()
```

Bunu <b>bar plot</b> şəklində göstərin:

```python
# your code goes here
dfmonth = df.groupby('Month')['Order_Quantity'].sum()
dfmonth.plot.bar(rot=0, figsize=(13,5))
```

### Ən yuxarı **satış sayı**na sahib ölkə hansıdır ?

```python
# your code goes here
df.groupby('Country')['Order_Quantity'].sum().idxmax()
```

Hər ölkə üzrə satışları <b>bar plot</b> şəklində göstərin:

```python
# your code goes here
df.groupby('Country')['Order_Quantity'].sum().plot.bar(rot=0, y= 'Order_Quantity', figsize=(13,5))
```

### Satılmış hər bir məhsulun göstərildiyi list yaradın

```python
# your code goes here
df['Product'].value_counts().index.tolist()
```

10 ən çox satılmış məhsulun satış miqdarını göstərən **bar plot** qurun (best sellers):

```python
# your code goes here
dfbar3 = df.groupby('Product')['Order_Quantity'].sum().nlargest(10)
dfbar3.plot.bar(rot=0, figsize=(16,7), fontsize=7.5)
```

### `Unit_Cost` və `Unit_Price` arasında necə bir əlaqə görə bilirsiniz?

Bunu plot-larla (uyğun gördüyünüz tip) bildirərək izah edin

### `Order_Quantity` and `Profit` arasında necə bir əlaqə görə bilirsiniz?

### `Profit` per `Country` arasında necə bir əlaqə görə bilirsiniz?

Bunu plot-larla (məs. qruplaşdırılmış box plot) bildirərək izah edin

```python
# your code goes here
fig, [ax1, ax2] = plt.subplots(2, 1, figsize=(12,12))
country = df.groupby('Country')['Profit'].sum()
country.plot(kind='bar', x='Country', y='Profit', figsize = (12,7), fontsize=8, ax=ax1 )
country.plot(kind='line', x='Country', y='Profit', color='blue', fontsize=10, ax=ax2 )
plt.show()

# Ən az gəliri olan ölkə Fransa, Ən çox gəliri olan ölkə Amerika olduğunu görmək olar 
```
